```python
class Kiosk:
    def __init__(self):
        # 메뉴와 가격을 딕셔너리로 통합 관리 (더 안전하고 직관적입니다)
        self.menu_data = {
            'americano': 2000,
            'latte': 3000,
            'mocha': 3000,
            'yuza_tea': 2500,
            'green_tea': 2500,
            'choco_latte': 3000
        }
        self.menu_list = list(self.menu_data.keys())
        self.price_sum = 0
        self.order_list = []

    def menu_print(self):
        print("\n--- [MENU] ---")
        for i, name in enumerate(self.menu_list, 1):
            print(f"{i}. {name:12} : {self.menu_data[name]}원")

    def _get_valid_input(self, prompt, valid_range):
        """숫자 입력을 안전하게 받는 헬퍼 함수"""
        while True:
            try:
                val = int(input(prompt))
                if val in valid_range:
                    return val
                print(f"{valid_range} 중 하나를 입력해주세요.")
            except ValueError:
                print("숫자만 입력 가능합니다.")

    def menu_select(self):
        while True:
            # 1. 메뉴 선택
            n = self._get_valid_input("\n음료 번호를 입력하세요: ", range(1, len(self.menu_list) + 1))
            selected_name = self.menu_list[n-1]
            
            # 2. 온도 선택
            t = self._get_valid_input("HOT(1) 또는 ICE(2)를 선택하세요: ", [1, 2])
            temp = "HOT" if t == 1 else "ICE"
            
            # 3. 주문 정보 저장
            price = self.menu_data[selected_name]
            self.price_sum += price
            order_item = f"{temp} {selected_name}"
            self.order_list.append(order_item)
            print(f"-> 추가됨: {order_item} ({price}원)")

            # 4. 추가 주문 여부
            again = self._get_valid_input("추가 주문하시겠습니까? (예:1, 아니오:2): ", [1, 2])
            if again == 2:
                break

    def pay(self):
        if not self.order_list:
            print("주문 내역이 없습니다.")
            return

        print(f"\n총 결제 금액: {self.price_sum}원")
        self._get_valid_input("결제 수단 선택 - 현금(1), 카드(2): ", [1, 2])
        
        print("결제가 진행 중입니다...")
        self.print_receipt()

    def print_receipt(self):
        print('\n' + '━' * 30)
        print("      [ 영 수 증 ]")
        print('━' * 30)
        for item in self.order_list:
            print(f" - {item}")
        print('-' * 30)
        print(f" 합계 금액: {self.price_sum}원")
        print('━' * 30)
        print(" 이용해주셔서 감사합니다! \n")
        
```

```python
# 여러 개 주문 기능과 예외 처리가 포함된 키오스크 실행
kiosk = Kiosk()
kiosk.menu_print()
kiosk.menu_select()   # 고의로 이상한 값을 입력해 보셔도 괜찮아요!
kiosk.pay()
```

```python
# 문제 1. 평행사변형과 사다리꼴의 넓이를 구하는 메서드 (유효성 검사 추가)

def par():
    try:
        width, vertical = map(float, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        if width <= 0 or vertical <= 0:
            return '오류: 가로와 세로는 0보다 커야 합니다.'
        area = width * vertical
        return f'평행사변형의 넓이는: {area}'
    except ValueError:
        return '오류: 숫자 형식이 올바르지 않습니다. (예: 3,4)'

def trape():
    try:
        low, upper, height = map(float, input('윗변, 아랫변, 높이를 입력하세요. 예시 : 윗변,아랫변,높이\n >>>').split(','))
        if low <= 0 or upper <= 0 or height <= 0:
            return '오류: 모든 길이는 0보다 커야 합니다.'
        area = (low + upper) * height / 2
        return f'사다리꼴의 넓이는: {area}'
    except ValueError:
        return '오류: 숫자 형식이 올바르지 않습니다. (예: 1,2,3)'
```

```python
# 1. 평행사변형 넓이 테스트
print("--- 평행사변형 넓이 구하기 ---")
print(par())

# 2. 사다리꼴 넓이 테스트
print("\n--- 사다리꼴 넓이 구하기 ---")
print(trape())
```
