## 문제 1. 최댓값과 최솟값 찾기



조건:  
* 주어진 리스트의 최댓값과 최솟값을 반환하는 클로저 함수를 완성해 주세요.
    * 주어진 숫자 리스트 numbers에 대해 find_min_max 함수를 호출하면 find_min과 find_max 함수를 얻을 수 있습니다.
    * find_min() 함수를 호출하면 최솟값을, find_max()함수를 호출하면 최댓값을 반환합니다.
    * 클로저를 사용하여 min_value와 max_value 변수를 외부에 노출하지 않고도 최솟값과 최대값을 찾을 수 있습니다.
    * find_min_max 함수가 호출된 이후에도 find_min과 find_max 함수를 통해 min_value와 max_value 변수에 접근할 수 있습니다.
* 주석으로 구현 내용이 써있는 부분을 코드로 채워주세요.
---
입력:
```python
numbers = [10, 5, 8, 12, 3, 7]
find_min, find_max = find_min_max(numbers)

print("최솟값:", find_min())  
print("최댓값:", find_max())  
```
---
출력:
```
3
12
```


````python
def find_min_max(numbers):
    # min_value와 max_value 변수를 초기화
    min_value = float('inf')
    max_value = float('-inf')

    def update_min_max(num):
        # 외부함수의 변수인 min_value, max_value 참조
        nonlocal min_value, max_value

        # 만약 num 값이 min_value보다 작다면 min_value를 num 값으로 변경
        if num < min_value:
            min_value = num

        # 만약 num 값이 max_value보다 크다면 max_value를 num 값으로 변경
        if num > max_value:
            max_value = num

    # numbers 리스트의 모든 값을 순환하며 최댓값과 최솟값 업데이트
    for num in numbers:
        update_min_max(num)

    # 최솟값을 반환하는 내부함수
    def get_min():
        return min_value

    # 최댓값을 반환하는 내부함수
    def get_max():
        return max_value

    # 외부함수는 내부함수(get_min()과 get_max())를 반환
    return get_min, get_max
````

````python
numbers = [10, 5, 8, 12, 3, 7]
find_min, find_max = find_min_max(numbers)

print("최솟값:", find_min())  # 3
print("최댓값:", find_max())  # 12
````

## 문제 2. 함수의 호출 횟수를 세는 데코레이터

지난 시간에 배운 데코레이터를 기억하시나요? 데코레이터도 내부함수를 반환하는 클로저를 활용한 문법입니다. 클로저를 활용하여 함수 호출 횟수를 출력하는 데코레이터를 만들어봅시다.

조건:
* 함수가 실행될 때 마다 함수 이름과 실행 횟수를 출력하는 데코레이터 `counter` 만들기  

힌트:
* 데코레이터의 내부 함수(클로저)는 실행횟수를 기억하기 위해 외부 함수의 변수를 참조 해야 합니다
* 함수의 이름은 `함수이름.__name__`으로 접근할 수 있습니다.

---
입력:
```python
@counter
def say_hello():
    print("Hello Aiffel!")

for i in range(5):
    say_hello()
```
---
출력:
```
Hello Aiffel!
say_hello 실행횟수: 1
Hello Aiffel!
say_hello 실행횟수: 2
Hello Aiffel!
say_hello 실행횟수: 3
Hello Aiffel!
say_hello 실행횟수: 4
Hello Aiffel!
say_hello 실행횟수: 5
```

````python
def counter(func):
    count = 0
    def wrapper(*args, **kwargs):
        # 실행 횟수를 기억하기 위해 외부 함수의 변수 참조
        nonlocal count
        count += 1
        result = func(*args, **kwargs)
        # 함수 이름과 실행 횟수 출력
        print(f"{func.__name__} 실행횟수: {count}")
        return result
    return wrapper
````

````python
@counter
def say_hello():
    print("Hello Aiffel!")

for i in range(5):
    say_hello()
````

클로저와 `nonlocal` 키워드의 개념을 더 명확히 이해하실 수 있도록 설명해 드릴게요.

### 1. 클로저 (Closure) 개념 설명

클로저는 파이썬에서 **'함수가 다른 함수를 반환할 때, 외부 함수의 스코프(환경)에 있는 변수들을 함께 기억하는'** 것을 의미합니다. 쉽게 말해, 반환된 내부 함수가 자신이 정의될 때의 환경을 "기억"하고 있기 때문에, 외부 함수가 이미 실행을 마치고 스코프를 벗어났음에도 불구하고 그 환경의 변수들에 접근할 수 있는 것입니다.

**간단한 예시:**

```python
def outer_function(msg):
    # 외부 함수의 변수 (자유 변수)
    message = msg

    def inner_function():
        # 내부 함수는 외부 함수의 message 변수를 기억하고 사용합니다.
        print(message)
    return inner_function

# outer_function을 호출하면 inner_function이 반환됩니다.
# 이때 inner_function은 message="Hello"를 기억합니다.
greet_hello = outer_function("Hello")
# outer_function은 이미 실행이 끝났지만, greet_hello는 message를 기억합니다.
greet_hello() # 출력: Hello

# 다른 message로도 클로저를 만들 수 있습니다.
greet_aiffel = outer_function("Aiffel")
greet_aiffel() # 출력: Aiffel
```
위 예시에서 `greet_hello`와 `greet_aiffel`은 `outer_function`이 반환한 `inner_function`입니다. 이 `inner_function`들은 `outer_function`이 종료된 후에도 각자 생성될 때의 `message` 값을 "기억"하고 있습니다. 이것이 바로 클로저의 핵심입니다.

### 2. `nonlocal` 키워드 설명

`nonlocal` 키워드는 **중첩된 함수 내에서 바로 바깥 함수의 (전역 변수도 지역 변수도 아닌) 변수를 '수정'할 수 있게 해주는** 역할을 합니다. 클로저가 외부 변수를 '기억'하는 것을 넘어서, `nonlocal`을 사용하면 이 기억된 변수의 값을 '변경'할 수 있게 되는 것이죠.

**간단한 예시 (숫자 세기):**

```python
def counter_factory():
    count = 0 # 바깥 함수의 변수

    def increment_counter():
        # nonlocal을 사용하여 바깥 함수의 count 변수를 수정합니다.
        nonlocal count
        count += 1
        return count
    return increment_counter

# 카운터 생성
my_counter = counter_factory()

print(my_counter()) # 출력: 1 (count가 0 -> 1로 변경)
print(my_counter()) # 출력: 2 (count가 1 -> 2로 변경)
print(my_counter()) # 출력: 3 (count가 2 -> 3로 변경)
```
이 예시에서 `increment_counter` 함수는 `nonlocal count`를 선언함으로써, 바로 바깥 함수인 `counter_factory`의 `count` 변수에 직접 접근하여 그 값을 수정할 수 있게 됩니다. `my_counter`를 호출할 때마다 `count` 값이 계속 증가하는 것을 볼 수 있습니다. `nonlocal`이 없었다면 `increment_counter` 내부에 새로운 지역 변수 `count`가 생성되어 바깥 `count`는 수정되지 않았을 겁니다.

### 3. `find_min_max`에 클로저와 `nonlocal` 적용 설명

이제 `find_min_max` 함수에서 이 개념들이 어떻게 사용되는지 살펴보겠습니다.

```python
def find_min_max(numbers):
    # 1. 외부 함수의 변수 초기화 (클로저가 '기억'할 대상)
    min_value = float('inf')
    max_value = float('-inf')

    def update_min_max(num):
        # 2. nonlocal을 사용하여 외부 함수의 min_value, max_value를 '수정'
        nonlocal min_value, max_value

        if num < min_value:
            min_value = num
        if num > max_value:
            max_value = num

    # 3. 모든 숫자에 대해 min_value와 max_value를 업데이트
    for num in numbers:
        update_min_max(num)

    # 4. 최솟값을 반환하는 내부 함수 (클로저의 한 부분)
    def get_min():
        # get_min은 외부 함수의 min_value를 '기억'하여 반환합니다.
        return min_value

    # 5. 최댓값을 반환하는 내부 함수 (클로저의 또 다른 부분)
    def get_max():
        # get_max는 외부 함수의 max_value를 '기억'하여 반환합니다.
        return max_value

    # 6. 외부 함수는 두 내부 함수를 반환합니다.
    # 이 두 함수는 find_min_max가 끝난 후에도 min_value와 max_value를 기억합니다.
    return get_min, get_max
```

*   **클로저의 역할**: `find_min_max` 함수는 `get_min`과 `get_max`라는 두 개의 내부 함수를 반환합니다. 이 두 함수는 `find_min_max`가 실행을 마친 후에도, 그 스코프에 정의되어 있던 `min_value`와 `max_value` 변수들을 "기억"하고 있습니다. 따라서 `find_min, find_max = find_min_max(numbers)`를 통해 반환된 `find_min()`과 `find_max()`를 호출하면, `find_min_max` 함수가 계산했던 최종 `min_value`와 `max_value`에 접근하여 그 값을 반환할 수 있는 것입니다.

*   **`nonlocal`의 역할**: `update_min_max` 함수는 `find_min_max` 내부의 `min_value`와 `max_value` 변수를 수정해야 합니다. 이때 `nonlocal min_value, max_value`를 선언함으로써, `update_min_max`가 지역 변수를 새로 만드는 대신, 바로 바깥 함수인 `find_min_max`의 `min_value`와 `max_value`를 직접 찾아 그 값을 업데이트할 수 있게 됩니다. 이 과정을 통해 `numbers` 리스트를 순회하면서 `min_value`와 `max_value`가 계속해서 최신 최솟값/최댓값으로 갱신됩니다.

결론적으로, `find_min_max` 함수는 `nonlocal`을 사용하여 `min_value`와 `max_value`를 지속적으로 업데이트하고, 클로저를 통해 이 최종 업데이트된 `min_value`와 `max_value`를 `get_min`과 `get_max` 함수가 기억하여 반환하도록 하는 멋진 디자인 패턴을 보여줍니다.

## `find_min_max`에 클로저와 `nonlocal` 적용 설명

앞서 배운 클로저와 `nonlocal` 개념이 `find_min_max` 함수에서 `min_value`와 `max_value`를 어떻게 '기억'하고 '수정'하는 데 사용되는지 단계별로 자세히 설명합니다.

#### 설명:

1.  **`min_value`와 `max_value` 변수 초기화**: `find_min_max(numbers)` 함수가 호출되면, 먼저 `min_value = float('inf')`와 `max_value = float('-inf')`로 두 변수를 초기화합니다. 이 변수들은 `find_min_max` 함수의 지역 변수이며, 나중에 반환될 내부 함수(`get_min`, `get_max`)들이 '기억'하게 될 대상이 됩니다. 즉, 클로저의 환경에 포함될 핵심 데이터입니다.

2.  **`update_min_max` 함수와 `nonlocal`**: `find_min_max` 함수 내부에 정의된 `update_min_max(num)` 함수는 `numbers` 리스트의 각 숫자를 처리하며 `min_value`와 `max_value`를 업데이트하는 역할을 합니다. 이때 `nonlocal min_value, max_value` 키워드를 사용합니다. 이 `nonlocal` 키워드는 `update_min_max` 함수가 자기 자신의 지역 변수를 만드는 대신, 바로 바깥쪽 함수(여기서는 `find_min_max`)의 `min_value`와 `max_value` 변수에 직접 접근하여 그 값을 '수정'할 수 있도록 합니다. `nonlocal`이 없었다면 `update_min_max` 내부에서 새로운 지역 변수가 생성되어 바깥 함수의 변수는 변경되지 않았을 것입니다.

3.  **`for` 루프를 통한 값 업데이트**: `find_min_max` 함수는 전달받은 `numbers` 리스트를 `for num in numbers:` 루프를 통해 순회합니다. 루프의 각 단계마다 `update_min_max(num)` 함수를 호출하여 현재 `num` 값에 따라 `find_min_max` 스코프의 `min_value`와 `max_value`를 최신 최솟값 및 최댓값으로 계속해서 '수정'합니다.

4.  **`get_min`, `get_max` 함수와 클로저**: `find_min_max` 함수는 `get_min()`과 `get_max()`라는 두 개의 내부 함수를 정의합니다. 이 함수들은 `return min_value`와 `return max_value`를 통해 `find_min_max` 함수의 최종 `min_value`와 `max_value` 값을 반환합니다. 여기서 중요한 점은 `find_min_max` 함수가 실행을 마쳐도, 이 두 내부 함수는 자신들이 정의될 당시의 환경, 즉 `min_value`와 `max_value` 변수들을 '기억'하고 있다는 것입니다. 이것이 바로 클로저의 핵심 원리입니다.

5.  **클로저로서의 반환**: 마지막으로 `find_min_max` 함수는 `return get_min, get_max`를 통해 이 두 내부 함수를 튜플 형태로 반환합니다. `find_min, find_max = find_min_max(numbers)`와 같이 호출하면, `find_min`과 `find_max` 변수에 각각 `get_min`과 `get_max` 함수가 할당됩니다. 이후 `find_min()`이나 `find_max()`를 호출할 때마다, `find_min_max` 함수가 이미 종료되었음에도 불구하고 클로저 덕분에 계산된 `min_value`와 `max_value`에 접근하여 원하는 값을 얻을 수 있게 됩니다. 이를 통해 외부에서는 내부 변수에 직접 접근하지 않고도 최솟값과 최댓값을 안전하게 조회할 수 있습니다.

````python
def find_min_max_with_trace(numbers):
    min_value = float('inf')
    max_value = float('-inf')

    print(f"--- 초기값: min_value={min_value}, max_value={max_value} ---")

    def update_min_max(num):
        nonlocal min_value, max_value
        print(f"\n현재 숫자: {num}")
        print(f"  업데이트 전: min_value={min_value}, max_value={max_value}")

        if num < min_value:
            print(f"  {num}이 현재 min_value({min_value})보다 작으므로 min_value 업데이트.")
            min_value = num
        else:
            print(f"  {num}이 현재 min_value({min_value})보다 작지 않음.")

        if num > max_value:
            print(f"  {num}이 현재 max_value({max_value})보다 크므로 max_value 업데이트.")
            max_value = num
        else:
            print(f"  {num}이 현재 max_value({max_value})보다 크지 않음.")
        print(f"  업데이트 후: min_value={min_value}, max_value={max_value}")

    for num in numbers:
        update_min_max(num)

    def get_min():
        return min_value

    def get_max():
        return max_value

    return get_min, get_max

numbers = [10, 5, 8, 12, 3, 7]
find_min_traced, find_max_traced = find_min_max_with_trace(numbers)

print(f"\n최종 최솟값: {find_min_traced()}")
print(f"최종 최댓값: {find_max_traced()}")
````
