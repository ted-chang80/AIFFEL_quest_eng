```python
# [정수 입력 강화된 꽃집 계산기]
import math

def calculator():
    print("--- [플라워486] 프리미엄 정산 계산기 ---")
    history = []  # 계산 내역을 저장할 바구니(리스트)

    while True:
        # 첫 번째 숫자 입력 (정수만 허용할 때까지 반복)
        while True:
            try:
                num1 = int(input("\n첫 번째 정수를 입력하세요: "))
                break  # 올바른 정수 입력 시 반복문 탈출
            except ValueError:
                # 숫자가 아니거나 소수점 포함 시 실행
                print("⚠️ 잘못된 입력입니다. 정수를 입력해주세요.")

        # 연산자 입력 받기
        operator = input("연산자 (+, -, *, /, **): ").strip()

        # 두 번째 숫자 입력 (정수만 허용할 때까지 반복)
        while True:
            try:
                num2 = int(input("두 번째 정수를 입력하세요: "))
                break
            except ValueError:
                print("⚠️ 잘못된 입력입니다. 정수를 입력해주세요.")

        try:
            result = None
            # 입력받은 연산자에 따라 계산 수행
            if operator == '+': result = num1 + num2
            elif operator == '-': result = num1 - num2
            elif operator == '*': result = num1 * num2
            elif operator == '/': result = num1 / num2
            elif operator == '**':
                # math.pow는 실수를 반환하므로 다시 정수(int)로 변환
                result = int(math.pow(num1, num2))

            if result is not None:
                output = f"{num1} {operator} {num2} = {result}"
                print(f"💰 결과: {output}")
                history.append(output)  # 계산 결과를 기록장에 추가
            else:
                print(f"⚠️ 지원하지 않는 연산자 '{operator}'입니다.")

        except ZeroDivisionError:
            # 0으로 나누려고 할 때 발생하는 오류 방지
            print("⚠️ 0으로 나눌 수 없습니다!")

        # 계속 계산할지 여부 확인
        while True:
            choice = input("\n계속할까요? (y/n): ").lower().strip()
            if choice in ['y', 'n']: break
            print("'y' 또는 'n'을 입력하세요.")

        # 종료 선택 시 기록을 출력하고 프로그램 종료
        if choice == 'n':
            print("\n--- [오늘의 정산 기록] ---")
            for item in history:
                print(f" - {item}")
            print("\n영업을 종료합니다. 감사합니다! 😊")
            break

if __name__ == '__main__':
    calculator()
```

### 🤖 꽃집 로봇과 코드의 비밀 연결 고리

| 로봇의 행동 (비유) | 실제 파이썬 코드 | 하는 일 |
| :--- | :--- | :--- |
| **1. 가게 문 열기** | `while True:` | 손님이 그만하라고 할 때까지 계속 반복해요. |
| **2. 숫자 물어보기** | `int(input(...))` | 손님이 말한 숫자를 로봇의 머릿속에 저장해요. |
| **3. "다시 말해줘!"** | `except ValueError:` | 글자를 입력하는 실수를 하면 친절하게 다시 물어봐요. |
| **4. 포장법 고르기** | `if operator == '+':` | 더하기, 빼기 등 어떤 계산을 할지 결정해요. |
| **5. 비밀 장부에 적기** | `history.append(...)` | 나중에 한꺼번에 보여주려고 계산 결과를 적어둬요. |
| **6. 0은 안 돼요!** | `except ZeroDivisionError:` | 수학적으로 불가능한 계산(0으로 나누기)을 막아줘요. |
| **7. 퇴근 인사** | `if choice == 'n': break` | 'n'을 누르면 장부를 보여주고 로봇이 잠들어요. |
