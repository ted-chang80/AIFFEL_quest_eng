```python
# detector를 선언합니다
detector_hog = dlib.get_frontal_face_detector() # 기본 얼굴 감지기를 반환
print("🌫🛸")
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)   # (image, num of image pyramid)
print("🌫🛸")
```

```python
# 찾은 얼굴 영역 박스 리스트
# 여러 얼굴이 있을 수 있습니다
print(dlib_rects)

for dlib_rect in dlib_rects: # 찾은 얼굴 영역의 좌표
    l = dlib_rect.left() # 왼쪽
    t = dlib_rect.top() # 위쪽
    r = dlib_rect.right() # 오른쪽
    b = dlib_rect.bottom() # 아래쪽

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) # 시작점의 좌표와 종료점 좌표로 직각 사각형을 그림

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
model_path = os.path.join(home_dir, 'work/camera_sticker/models/shape_predictor_68_face_landmarks.dat')
    # 저장한 landmark 모델의 주소를 model_path 변수에 저장
landmark_predictor = dlib.shape_predictor(model_path)
    # dlib 라이브러리의 shape_predictor 함수를 이용하여 모델을 불러옴
    # landmark_predictor는 RGB이미지와 dlib.rectangle를 입력 받고 dlib.full_object_detection를 반환
    # dlib.rectangle: 내부를 예측하는 박스
    # dlib.full_object_detection: 각 구성 요소의 위치와, 이미지 상의 객체의 위치를 나타냄
print("🌫🛸")
```

```python
list_landmarks = []
    # 랜드마크의 위치를 저장할 list 생성

# 얼굴 영역 박스 마다 face landmark를 찾아냅니다
# face landmark 좌표를 저장해둡니다
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
        # 모든 landmark의 위치정보를 points 변수에 저장
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        # 각각의 landmark 위치정보를 (x,y) 형태로 변환하여 list_points 리스트로 저장
    list_landmarks.append(list_points)
        # list_landmarks에 랜드마크 리스트를 저장

print(len(list_landmarks[0]))
    # 얼굴이 n개인 경우 list_landmarks는 n개의 원소를 갖고
    # 각 원소는 68개의 랜드마크 위치가 나열된 list
    # list_landmarks의 원소가 1개이므로 list_landmarks[1]을 호출하면 IndexError가 발생
```

```python
# list_landmarks의 원소가 1개이므로 아래 반복문은 한번만 실행됨
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 5, (0, 255, 255), -1)
            # cv2.circle: OpenCV의 원을 그리는 함수
            # img_show 이미지 위 각각의 point에
            # 크기가 5이고 (0, 255, 255)색으로 내부가 채워진(-1) 원을 그림
            # (마지막 인수가 자연수라면 그만큼의 두께의 선으로 원이 그려짐)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
    # RGB 이미지로 전환
plt.imshow(img_show_rgb)
    # 이미지를 준비
plt.show()
    # 이미지를 출력
```

```python
# x, y는 코의 중심 좌표, w, h는 스티커의 너비와 높이
# 스티커의 시작점(top-left)을 계산하여 코의 중심에 오도록 조정합니다.
refined_x = x - w // 2 # 스티커의 시작 X좌표 (코 x - 스티커 너비의 절반)
refined_y = y # 스티커의 시작 Y좌표 (코 y)

# 원본 이미지에 스티커 이미지를 추가하기 위해서 x, y 좌표를 조정합니다. 이미지 시작점은 top-left 좌표이기 때문입니다.
print (f'(x,y) : ({refined_x},{refined_y})')
# opencv는 ndarray데이터를 사용하는데, ndarray는 음수인덱스에 접근 불가하므로 스티커 이미지를 잘라 줘야 한다.
```

```python
# 1. 깔끔한 시작을 위해 원본 이미지를 다시 로드합니다.
my_image_path = os.path.join(home_dir, 'work/camera_sticker/images/image.png')
img_bgr = cv2.imread(my_image_path) # 원본 BGR 이미지를 로드합니다.
img_show = img_bgr.copy() # 결과 표시를 위해 사본을 생성합니다.
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB) # dlib을 위해 RGB로 변환합니다.
```

```python
# 2. 얼굴과 랜드마크를 다시 감지합니다.
dlib_rects = detector_hog(img_rgb, 1)
list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    # 모든 landmark의 위치정보를 points 변수에 저장
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    # 각각의 landmark 위치정보를 (x,y) 형태로 변환하여 list_points 리스트로 저장
    list_landmarks.append(list_points)
     # list_landmarks에 랜드마크 리스트를 저장
```

```python
# 스티커 위치 및 크기 조절 파라미터
# y_offset: 코 랜드마크를 기준으로 스티커를 위/아래로 이동 (음수: 위로, 양수: 아래로)
# sticker_scale: 스티커의 크기 조절 (1.0 = 원본 크기, 0.5 = 절반 크기, 2.0 = 두 배 크기)
y_offset = -50
sticker_scale = 1.0
```

```python
# 3. 스티커 위치 및 초기 크기 계산 (예: 코를 중심으로)
# 단순화를 위해 하나의 얼굴만 가정하며, 첫 번째 감지된 얼굴을 사용합니다.
if dlib_rects and list_landmarks:
    dlib_rect = dlib_rects[0]
    landmark = list_landmarks[0]

    x = landmark[33][0] # 코의 x 좌표
    y = landmark[33][1] # 코의 y 좌표

    # 스티커 스케일 적용
    w = int(dlib_rect.width() * sticker_scale) # 얼굴 너비를 기반으로 한 스티커 너비에 스케일 적용
    h = int(dlib_rect.height() * sticker_scale) # 얼굴 높이를 기반으로 한 스티커 높이에 스케일 적용

    # 계산된 w, h에 맞춰 스티커 이미지 크기 조정
    global img_sticker # 이전에 정의된 전역 img_sticker를 사용합니다.
    sticker_path = '/content/cat-whiskers-removebg-preview.png' # 마지막으로 선택된 스티커를 사용합니다.
    img_sticker = cv2.imread(sticker_path, cv2.IMREAD_UNCHANGED) # 알파 채널과 함께 로드합니다.
    if img_sticker is None:
        print(f"오류: {sticker_path} 에서 스티커 이미지를 로드하지 못했습니다.")
    else:
        img_sticker = cv2.resize(img_sticker, (w, h), interpolation=cv2.INTER_AREA)

    # 코를 중심으로 스티커의 좌상단 모서리 계산 및 y_offset 적용
    refined_x = x - w // 2
    refined_y = y - h // 2 + y_offset # y_offset을 더하여 수직 위치 조절
```

```python
# 4. 이미지 경계를 벗어나는 스티커 좌표 처리
img_h, img_w, _ = img_show.shape
sticker_h, sticker_w = img_sticker.shape[0], img_sticker.shape[1] # 크기 조정 후 실제 스티커 크기 가져오기

# 스티커 이미지의 자르기 좌표 초기화
sticker_crop_y_start = 0
sticker_crop_x_start = 0
sticker_crop_y_end = sticker_h
sticker_crop_x_end = sticker_w

# 상단 및 좌측 경계 조정 (음수 좌표)
if refined_y < 0:
    sticker_crop_y_start = -refined_y
    refined_y = 0

if refined_x < 0:
    sticker_crop_x_start = -refined_x
    refined_x = 0

# 하단 및 우측 경계 조정 (스티커가 이미지를 벗어나는 경우)
if refined_y + sticker_h > img_h:
    sticker_crop_y_end = img_h - refined_y

if refined_x + sticker_w > img_w:
    sticker_crop_x_end = img_w - refined_x

# 계산된 경계를 기반으로 스티커 자르기
img_sticker_cropped = img_sticker[sticker_crop_y_start:sticker_crop_y_end, sticker_crop_x_start:sticker_crop_x_end]

# 메인 이미지에 붙여넣을 최종 크기 결정
paste_h = img_sticker_cropped.shape[0]
paste_w = img_sticker_cropped.shape[1]

print(f"스티커 배치 좌표 (refined_x, refined_y): ({refined_x}, {refined_y})")
print(f"스티커 잘린 크기 (paste_w, paste_h): ({paste_w}, {paste_h})")
```

```python
# 5. 잘린 스티커를 이미지에 적용합니다.
# 필요한 경우 img_sticker_cropped에 알파 채널이 있는지 확인하거나 투명도를 별도로 처리합니다.
# 단순화를 위해 3개 채널을 가정하고, 검은색(0)이 아닌 경우 np.where를 사용하여 적용합니다.

# 스티커를 적용할 기본 이미지 영역을 추출합니다.
sticker_area_on_img = img_show[refined_y : refined_y + paste_h, refined_x : refined_x + paste_w]

# 스티커를 적용합니다. 스티커에 알파 채널이 있는 경우, 더 정교한 블렌딩이 필요합니다.
# 현재는 알파가 명시적으로 처리되지 않으면 투명 영역은 검은색(0,0,0)으로 가정합니다.
# 스티커에 알파 채널이 있는 경우, BGR 및 알파 마스크로 변환합니다.
if img_sticker_cropped.shape[2] == 4: # 알파 채널이 있는지 확인합니다.
    bgr = img_sticker_cropped[:, :, :3]  # BGR 채널
    alpha = img_sticker_cropped[:, :, 3] / 255.0  # 부동 소수점 [0, 1] 범위의 알파 채널

    # 스티커와 배경을 블렌딩합니다.
    for c in range(0, 3):
        sticker_area_on_img[:, :, c] = sticker_area_on_img[:, :, c] * (1 - alpha) + bgr[:, :, c] * alpha
    img_show[refined_y : refined_y + paste_h, refined_x : refined_x + paste_w] = sticker_area_on_img
else: # 알파 채널이 없는 경우, 검은색이 아닌 픽셀에 대해 np.where를 사용합니다.
    img_show[refined_y : refined_y + paste_h, refined_x : refined_x + paste_w] = \
        np.where(img_sticker_cropped == 0, sticker_area_on_img, img_sticker_cropped).astype(np.uint8)
```

```python
# 고양이 수염 이미지를 적용한 이미지를 보여준다.
# 얼굴 영역과 랜드마크를 미리 적용해놓은 img_show에 고양이 수염 이미지를 덧붙인 이미지가 나오게 된다.)
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```
